```python
Name = input('Name')
Eng = int(input ('English'))
Urdu = int(input ('Urdu'))
Math = int(input ('Math'))
Computer = int(input ('Computer'))
Sci = int(input ('Science'))
PST = int(input ('Pakistan Studies'))
Total= Eng+Urdu+Math+Computer+Sci+PST
print(Total)
Perc = Total/600*100
print(Perc)


if Perc > 100:
    print('Cheater')
elif Perc >= 80  :
    print('A+')
elif Perc >= 70:
    print('A')
elif Perc >= 60:
    print('B')
elif Perc >= 50:
    print('C')
elif Perc < 50:
    print('Fail')
else:
    print('Error')
    
```
