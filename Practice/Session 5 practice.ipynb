```python
Quiz = [
    {
    "S No." : 1,
    "question" : "What is your country name",
    "options": ["Pakistan","India","China","Iran"],
    "answer": "Pakistan"
    },
    {
    "S No." : 2,
    "question" : "What is your city name",
    "options": ["Karachi","Islamabad","Quetta","Sukkur"],
    "answer": "Karachi"
           
    },
    {
    "S No." : 3,
    "question" : "What is your  name",
    "options": ["Ali","Amjad","Adnan","Anjum"],
    "answer": "Ali"
    },
    {
    "S No." : 4,
    "question" : "What is your phone code",
    "options": ["454","91","54","+92"],
    "answer": "+92"
    }
]

```

```python
quiz
```

```python
Score = 0
for QuizApp in Quiz:
    Data = str(QuizApp['S No.']) +" "+ QuizApp['question']+"\n"+QuizApp['options'][0]+"\n"+QuizApp['options'][1]+"\n"+QuizApp['options'][2]+"\n"+QuizApp['options'][3]+"\n"
       
    Correct_Ans = input(Data)
    if Correct_Ans == QuizApp['answer']:
        Score += 1
print(Score) 
```
