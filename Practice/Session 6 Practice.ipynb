```python
def salary_calculator(Basic,*Alloance,**deduction):
    Total_allowance = sum(Alloance)
    Total_deduction = sum(deduction.values())

    Gross_Salary= Basic+Total_allowance
    Net_Salary= Gross_Salary-Total_deduction

    print('Base Salary',Basic)
    print('Total Allooance',Total_allowance)
    print('Total Deduction',Total_deduction)
    print('Net Salary',Net_Salary)

    return Net_Salary
salary_calculator(50000,2000,1000,Fine=2000,Leaves=1000)
```
