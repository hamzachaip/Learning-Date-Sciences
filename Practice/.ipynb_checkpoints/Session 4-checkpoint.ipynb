```python
val1 = int(input("Enter Value 1"))
val2 = int(input("Enter Value 2"))
op = input("Enter Operator")
if op == "+":
    print(val1 + val2)
elif op == "-":
    print(val1 - val2)
elif op == "x":
    print(val1 * val2)
elif op == "/":
    print(val1 / val2)
else:
    print("Please Enter Correct Operator")
```

```python
cleanCities = ["Karachi","Lahore","Sukkur","Multan","Quetta"]
cityName = input("Enter city Name:")
for cityToCheck in cleanCities:
    if cityName == cityToCheck:
        print("It is a clean city")
```

```python
firstName = ["Ali","Abid","Adnan"]
lastName = ["Ahmed","Anjum","Azad"]
for first in firstName:
    for last in lastName:
        print(first + " " + last)
```

```python
# While Loop

a = 0
while a < 5:
    print(a)
    a += 1 
```

```python
cities_to_visit = []
user_input = ''

while user_input != 'quit':
    city = input('Enter city to visit:')
    if city == 'quit':
        user_input = 'quit'
    else:
        cities_to_visit.append(city)
    
```

```python
print(cities_to_visit)
```
