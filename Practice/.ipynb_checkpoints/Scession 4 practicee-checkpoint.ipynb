```python
num1 = int(input('Enter your Number'))
num2 = int(input('Enter your second number'))
op = input('Enter your op')

if op == '+':
    print(num1 + num2)
elif op == '-':
    print(num1 - num2)
elif op == '*':
    print (num1 * num2)
elif op == '/':
    print (num1 / num2)
```

```python
# Nested Loop
Fname=['Hamza','Daniyal','Zafar']
Std_names=['Hamza','Ali','Sunny','Annes','Fazan']

for Student in Std_names:
    for Fathername in Fname:
        print (Student + " " + Fathername)
```
