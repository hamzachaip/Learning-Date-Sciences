```python
math=int(input("Enter Maths Marks"))
urdu=int(input("Enter Urdu Marks"))
eng=int(input("Enter English Marks"))
isl=int(input("Enter Islamiat Marks"))
sin=int(input("Enter Sindhi Marks"))
obtain_Marks=math+urdu+eng+isl+sin
perc = obtain_Marks/500*100
print("Your Percentage is:",perc)
if perc >=80 and perc <= 100:
    print("You have secured A+ Grade")
elif perc >=70 and perc < 80:
    print("You have secured A Grade")
elif perc >=60 and perc < 70:
    print("You have secured B Grade")
elif perc >=50 and perc < 60:
    print("You have secured C Grade")
elif perc >=40 and perc < 50:
    print("You have secured D Grade")
elif perc >=33 and perc < 40:
    print("You have secured E Grade")
elif perc >=0 and perc < 33:
    print("You have secured F Grade")
else:
    print("Your Percentage is wrong")
```
